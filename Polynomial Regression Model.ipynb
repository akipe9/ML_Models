```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv(r'C:\Users\akiper\Desktop\Machine Learning Course\Machine-Learning-A-Z-New\Machine Learning A-Z New\Part 2 - Regression\Section 6 - Polynomial Regression\Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
#Training the Linear Regression Model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y) #Using the whole dataset because the dataset is so small, we want to utilize as much as possible
```

```python
#Training the Polynomial Regression Model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
#Visualizing the Linear Regression Results
plt.scatter(x, y, c='red')
plt.plot(x, lin_reg.predict(x), c='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing the Polynomial Regression Results
plt.scatter(x, y, c='red')
plt.plot(x, lin_reg_2.predict(x_poly), c='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show() #Note: This result with degree of 4 is definitely overfitting 
```

```python
#Predicting a new result (a specific value) with Linear Regression
lin_reg.predict([[6.5]])
```

```python
#Predicting a new result (a specific value) with Polynomial Regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
